```python
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy
import trimesh
```

```python
folder = 'E:/obj/'
extension = '.obj'

# recursive search for files with extension
files = [os.path.join(dp, f) for dp, dn, fn in os.walk(os.path.expanduser(folder)) for f in fn]
files = [f for f in files if f.endswith(extension)]
print('Number of meshes: %d' % len(files))
```

```python
# random mesh
random_index = np.random.randint(0, len(files))
mesh = trimesh.load_mesh(files[random_index])

print('Mesh: %s' % files[random_index])

# plot mesh
#mesh.apply_transform(trimesh.transformations.rotation_matrix(np.pi, [0, 0, 1]))
#mesh.apply_transform(trimesh.transformations.rotation_matrix(np.pi/8, [1, 0, 0]))
plt.savefig('results/mesh.png', bbox_inches='tight', pad_inches=0, dpi=500)
mesh.show()
```

```python
# retrieve points below 0.01 and above -0.01
points = mesh.vertices
points = points[(points[:, 2] < 0.01) & (points[:, 2] > -0.01) & (points[:, 0] > -0.01)]

# show points
color = np.arange(len(points))
plt.scatter(points[:, 0], points[:, 1], c='black', s=1)
# scale axis to fit points
plt.axis('scaled')
# remove axes' numbers
plt.xticks([])
plt.yticks([])
# save to disk
plt.savefig('results/points.png', bbox_inches='tight', pad_inches=0, dpi=500)
```
