```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_iterations = (25, 16)
num_fragment = (2, 10)
```

```python
# interpolate num_iterations over the range of num_fragment
num_generated_fragments = np.zeros(num_fragment[1] - num_fragment[0] + 1)
current_sum = .0

for n_frag in range(num_fragment[0], num_fragment[1] + 1):
    alpha = (n_frag - num_fragment[0]) / (num_fragment[1] - num_fragment[0])
    it = int(num_iterations[0] * (1 - alpha) + num_iterations[1] * alpha)
    if current_sum + it * n_frag > 1000:
        num_generated_fragments[n_frag - num_fragment[0]] = 1000 - current_sum
        break
    else:
        num_generated_fragments[n_frag - num_fragment[0]] = it * n_frag
        
    current_sum += it * n_frag

# plot
plt.plot(range(num_fragment[0], num_fragment[1] + 1), num_generated_fragments)
plt.xlabel('Number of fragments')
plt.ylabel('Number of fragments * number of iterations')
plt.show()
```
